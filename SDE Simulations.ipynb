```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import standard_normal
from scipy.interpolate import interp1d
from scipy.stats import poisson
```

# Generate Wiener process

```python
# Parameters
T = 1  # total time
N = 100  # number of time steps
dt = T/N  # time step
mu = 0.4  # mean
sigma = 1  # standard deviation
num_sims = 5  # number of simulations

# Generate Wiener process
np.random.seed(0)
for i in range(num_sims):
    W = np.random.normal(mu, sigma, N) * np.sqrt(dt)
    W = np.cumsum(W)
    W = np.insert(W, 0, 0)

    # Plot Wiener process
    plt.plot(np.linspace(0, T, N+1), W)

plt.xlabel('Time')
plt.ylabel('Wiener Process')
plt.title('Multiple Simulations of Wiener Process')
plt.show()
```

# Ornstein-Uhlenbeck process

```python
# Parameters
T = 1  # total time
N = 100  # number of time steps
dt = T/N  # time step
mu = 0.4  # mean
theta = 0.15 # damping coefficient
sigma = 0.2 # volatility coefficient
num_sims = 5  # number of simulations

# Generate Ornstein-Uhlenbeck process
np.random.seed(0)
for i in range(num_sims):
    X = np.zeros(N+1)
    for j in range(1, N+1):
        X[j] = X[j-1] + theta*(mu - X[j-1])*dt + sigma*np.sqrt(dt)*np.random.normal()
    # Plot Ornstein-Uhlenbeck process
    plt.plot(np.linspace(0, T, N+1), X)

plt.xlabel('Time')
plt.ylabel('Ornstein-Uhlenbeck Process')
plt.title('Multiple Simulations of Ornstein-Uhlenbeck Process')
plt.show()
```

# Geometric Brownian motion

```python
# Parameters
T = 1 # total time
N = 100 # number of time steps
dt = T/N # time step
mu = 0.4 # drift coefficient
sigma = 0.2 # volatility coefficient
S_0 = 1 # initial value
num_sims = 5 # number of simulations

# Generate Geometric Brownian motion
np.random.seed(0)
for i in range(num_sims):
    S = np.zeros(N+1)
    S[0] = S_0
    for j in range(1, N+1):
        S[j] = S[j-1]*np.exp((mu-0.5*sigma**2)*dt + sigma*np.sqrt(dt)*np.random.normal())
    # Plot Geometric Brownian motion
    plt.plot(np.linspace(0, T, N+1), S)

plt.xlabel('Time')
plt.ylabel('Geometric Brownian Motion')
plt.title('Multiple Simulations of Geometric Brownian Motion')
plt.show()
```

# Fractional Brownian motion

```python
# Parameters
T = 1 # total time
N = 100 # number of time steps
dt = T/N # time step
H = 0.15 # Hurst exponent
num_sims = 5 # number of simulations

# Generate Fractional Brownian motion
np.random.seed(0)
for i in range(num_sims):
    t = np.linspace(0, T, N+1)
    f = interp1d(t, standard_normal(N+1))
    x = np.cumsum(f(np.linspace(0, T, N**2+1)))
    x = np.interp(np.linspace(0, T, N+1), np.linspace(0, T, N**2+1), x)
    x = x - x[0]
    x = x/np.std(x)
    x = x*np.sqrt(dt)**H
    # Plot Fractional Brownian motion
    plt.plot(np.linspace(0, T, N+1), x)

plt.xlabel('Time')
plt.ylabel('Fractional Brownian Motion')
plt.title('Multiple Simulations of Fractional Brownian Motion')
plt.show()
```

# Poisson point process

```python
# Parameters
lam = 5 # parameter of the Poisson distribution
num_sims = 5 # number of simulations

# Generate Poisson point process
np.random.seed(0)
for i in range(num_sims):
    # Generate Poisson distributed points
    points = poisson.rvs(lam, size=(2, poisson.rvs(lam)))
    # Plot the points
    plt.scatter(points[0], points[1])

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Multiple Simulations of Poisson Point Process')
plt.show()
```
